### Criando uma categorização para a localização

```python
def categorizacao(local):
    if local < 100:
        return 'Generalidades. Ciência e conhecimento'
    elif local < 200:
        return 'Filosofia e psicologia'
    elif local < 300:
        return 'Religião'
    elif local < 400:
        return 'Ciências sociais'
    elif local < 500:
        return 'Classe vaga. Provisoriamente não ocupada'
    elif local < 600:
        return 'Matemática e ciências naturais'
    elif local < 700:
        return 'Ciências aplicadas'
    elif local < 800:
        return 'Belas artes'
    elif local < 900:
        return 'Linguagem. Língua. Linguística'
    elif local < 1000:
        return 'Geografia. Biografia. História.'
```

### Aplicanto a categorização no DataFrame

```python
dados_full['classe_CDU'] = dados_full['localizacao'].apply(categorizacao)
```

```python
dados_full.head()
```

### Transformando a coluna 'matricula_ou_siape' em string

```python
dados_full['matricula_ou_siape'] = dados_full['matricula_ou_siape'].astype('string')
```
