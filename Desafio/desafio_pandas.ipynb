```python
def categorizacao(local):
    if local < 100:
        return 'Generalidades. Ciência e conhecimento'
    elif local < 200:
        return 'Filosofia e psicologia'
    elif local < 300:
        return 'Religião'
    elif local < 400:
        return 'Ciências sociais'
    elif local < 500:
        return 'Classe vaga. Provisoriamente não ocupada'
    elif local < 600:
        return 'Matemática e ciências naturais'
    elif local < 700:
        return 'Ciências aplicadas'
    elif local < 800:
        return 'Belas artes'
    elif local < 900:
        return 'Linguagem. Língua. Linguística'
    elif local < 1000:
        return 'Geografia. Biografia. História.'
```

### Função para criar o boxplot

```python
def cria_box_plot(dataset,x,y,titulo,subtitulo):

  sns.set_theme(style="darkgrid", palette='Blues',font_scale=1.3)
  plt.figure(figsize=(16,10))

  ax = sns.boxplot(y= y, x= x, data= dataset,color='#4171EF')


  plt.ylim(0,max(dataset[y])*1.1) #Definir o limite do eixo y
  plt.xlabel(None)
  plt.ylabel(None)

  ax.set_title(titulo+"\n",size=20,loc='left',weight='bold')
  ax.text(s=subtitulo,x=-0.5,y=max(dataset[y])*1.11,fontsize=18, ha='left',color='gray')
```

### Box_Plot de Alunos Graduação

```python
cria_box_plot(alunos_graduacao_high_frequency,'ano','quantidade','Distribuição dos empréstimos mensais', 'Realizados pelos alunos de graduação no acervo circulante')
```

### Box_Plot de Alunos Pós-Graduação

```python
cria_box_plot(alunos_pos_graduacao_high_frequency,'ano','quantidade','Distribuição dos empréstimos mensais', 'Realizados pelos alunos de pós graduação no acervo circulante')
```

```python
def diferenca_percentual_ano_anterior(x,y):
  return round(((x / y * 100) - 100),2)
```

### Criando a tabela em HTML 

```python
th_props = [
  ('font-size', '1.4rem'),
  ('text-align', 'center'),
  ('font-weight', 'bold'),
  ('color', 'whitesmoke'),
  ('background-color', '#001692'),
  ('border-radius', '0.25rem'),
  ('box-shadow','0 0 1rem gray')
  ]

td_props = [
  ('font-size', '1rem'),
  ('padding','0.5rem'),
  ('text-align', 'left'),
  ('font-weight', 'bold'),
  ('border-bottom','0.1rem solid lightgray')
  ]

styles = [
  dict(selector="th", props=th_props),
  dict(selector="td", props=td_props)
  ]
```

### Gerando a tabela

```python
percentual.style.text_gradient(cmap='RdYlGn',low=1, axis=1,vmax=0.1,vmin=0)\
                              .format('{:.2f} %',subset=['2018','2019','2022'])\
                              .hide_index()\
                              .set_table_styles(styles)\
                              .to_html('teste.html',doctype_html =True,
                                       table_attributes='ALIGN=LEFT WIDTH=50% CELLSPACING = 5')
```
